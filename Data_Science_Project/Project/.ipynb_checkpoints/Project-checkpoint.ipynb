In de volgende stap wordt Wifi_Info gesplitst zodanig dat de tabel genormaliseerd wordt.

```python
# Efficiently split Pandas Dataframe cells containing lists into multiple rows, duplicating the other column's values.
# Credit goes to https://gist.github.com/jlln/338b4b0b55bd6984f883#gistcomment-2359013
def splitDataFrameList(df,target_column,separator):
    row_accumulator = []

    def splitListToRows(row, separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)

    df.apply(splitListToRows, axis=1, args = (separator, ))
    new_df = pd.DataFrame(row_accumulator)
    return new_df

data = splitDataFrameList(data, 'Wifi_Info', ',')
data.info()
data.head()
```

Na de data uit Wifi_Info te normaliseren breidt de dataset uit naar 10137 rijen. In de volgende stap wordt de data gesplitst van de identifiers, zodanig dat deze data bruikbaar wordt in modellen.

```python
# Extracts and splits the location from the imported data file.
location = data['Location'].str.split('@', n = 1, expand = True)

# Extracts and splits information about the Wifi from the imported data file.
wifi_info = data['Wifi_Info'].str.split('@', n = 1, expand = True)
# Cleans up the SSID
for x in range(len(wifi_info[0])):
    wifi_info[0][x] = wifi_info[0][x].replace('WifiInfo=', '').strip()

data['Sender'] = data['Sender'].str.split('=', n = 1, expand = True)[1]
data['Location'] = location[0].str.split('=', n = 1, expand = True)[1]
data['X'] = location[1].str.split(';', n = 1, expand = True)[0].astype(float)
data['Y'] = location[1].str.split(';', n = 1, expand = True)[1].astype(float)
data['SSID'] = wifi_info[0]
data['BSSID'] = wifi_info[1].str.split(':', n = 1, expand = True)[0]
data['Signal_strength'] = wifi_info[1].str.split(':', n = 1, expand = True)[1].astype(int)

data.info()
data.head()
```

In het 'classification report' van sklearn worden verschillende gegevens meegegeven over het model. Zo is er informatie over de precision (TP/(TP + FP)), recall (TP/(TP+FN)), f1-score en support. Over het algemeen wordt best gekeken naar het 'weighted average' van de precision om de nauwkeurigheid van een model te evalueren. Dit veld staat voor het gewogen gemiddelde van _2 * (recall * precision) / (recall + precision)_.

```python
rmse_val = [] #to store rmse values for different k
for K in range(1,20):
    model = KNeighborsRegressor(n_neighbors = K, metric='manhattan', weights='distance')
    model.fit(X_train_coordinates_scaled, y_train_coordinates)  #fit the model
    pred = model.predict(X_test_coordinates_scaled) #make prediction on test set
    error = sqrt(mean_squared_error(y_test_coordinates, pred)) #calculate rmse
    rmse_val.append(error) #store rmse values

plt.figure(figsize=(12, 6))  
plt.plot(range(1, 20), rmse_val, color='red', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=10)
plt.xticks(range(1, 20))
plt.title('Error Rate K Value')  
plt.xlabel('K Value')  
plt.ylabel('Mean Error')  
```

Voor KNN wordt altijd een oneven k-waarde gekozen om in het geval van een gelijkstand om een _tiebreaker_ te hebben.  
Opmerkelijk is dat bij een grote k-waarde niet alleen de complexiteit zal stijgen, maar ook de mean error van het model.
